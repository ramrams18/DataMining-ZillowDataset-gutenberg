```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("M.csv", parse_dates=True, index_col='RegionName')
data.head()
```

```python
copied = data.copy(deep=False)
copied = copied.T
percentages = copied.copy(deep=False)
percentages = percentages[4:]
```

```python
date_list = []
for row in copied:
    date_list.append(row)
for d in date_list:
    percentages[d] = percentages[d].astype(float).pct_change(fill_method='ffill')
percentages.index = pd.to_datetime(percentages.index).strftime('%Y-%m')
percentages.index=percentages.index.to_series().astype(str)
percentages
```

```python
#Range and frame
four_r = pd.period_range("2000-01-01", "2004-01-01", freq='m')
four_f = percentages[:len(four_r)]
eight_r = pd.period_range("2000-01-01", "2008-01-01", freq='m')
eight_f = percentages[:len(eight_r)]
ten_r = pd.period_range("2000-01-01", "2010-01-01", freq='m')
ten_f = percentages[:len(ten_r)]
```

```python
four_agg = four_f.agg(['sum'])
four_sort = four_agg.sort_values(by=['sum'], axis=1, ascending=False)

eight_agg = eight_f.agg(['sum'])
eight_sort = eight_agg.sort_values(by=['sum'], axis=1, ascending=False)

ten_agg = ten_f.agg(['sum'])
ten_sort = ten_agg.sort_values(by=['sum'], axis=1, ascending=False)
```

```python
#parta price increase over time span( over 4, 8 and 10 years) of N best cities
n = input('Enter the N value:')
print('Over four years:')
print(four_sort.iloc[0][:int(n)])
print('Over eight years:')
print(eight_sort.iloc[0][:int(n)])
print('Over Ten years:')
print(ten_sort.iloc[0][:int(n)])
```

```python
#partB ook a city name and compared it’s performance (price increases) over 5 & 10 year spans to the overall averages for that state as well as for the US. 
date_list
five_r = pd.period_range("2000-01-01", "2005-01-01", freq='m')
five_f = percentages[:len(five_r)]
five_agg = five_f.agg(['sum'])
five_sort = five_agg.sort_values(by=['sum'], axis=1, ascending=False)
jyu_r = pd.period_range("2010-01-01", "2005-01-01", freq='m')
jyu_f = percentages[:len(jyu_r)]
jyu_agg = jyu_f.agg(['sum'])
jyu_sort = jyu_agg.sort_values(by=['sum'], axis=1, ascending=False)
```

```python
city = input('Enter the city name:')
match_name = ''
for name in date_list:
    if city.lower() in name.lower():
        match_name = name
        break
state = match_name[-2:]
print(match_name, state)
filter_names = filter(lambda name: name.endswith(state), date_list)
filter_names = list(filter_names)

print('Percentage increase over 5 years:')
print(match_name, five_sort[match_name])
sum = 0
for cityn in filter_names:
    sum = sum+five_sort[cityn]
print(state, ': ', sum)
print('United States: ', five_sort['United States'].iloc[0])
```

```python
#partc calculated top 10 markets over 5 years
years = input('Enter no of years')
end_year = 2000+int(years)

custom_range = pd.period_range("2000-01-01", f"{end_year}-01-01", freq='m')
custom_frame = percentages[:len(custom_range)]
custom_agg = custom_frame.agg(['sum'])
custom_sorted = custom_agg.sort_values(by=['sum'], axis=1, ascending=False)
print(f'Top 10 markets Over {years} years: ', custom_sorted.iloc[0][:10])
```
